```python
import numpy as np
```

```python
import pandas as ps
```

```python
import matplotlib.pyplot as malt 
```

```python
from sklearn.preprocessing import Normalizer
```

```python
norm_df=ps.DataFrame({'x':np.random.randint(-100,100,1000),
                      'y':np.random.randint(-80,80,1000),
                      'z':np.random.randint(-150,150,1000)})
```

```python
norm_df
```

```python
ax=malt.axes(projection='3d')
ax.scatter3D(norm_df.x,norm_df.y,norm_df.z)
malt.figure()
```

```python
normal=Normalizer()
```

```python
norm_df=normal.fit_transform(norm_df)
norm_df
```

```python
norm_df=ps.DataFrame(norm_df,columns=['x','y','z'])
norm_df
```

```python
ax=malt.axes(projection='3d')
ax.scatter3D(norm_df.x,norm_df.y,norm_df.z)
malt.figure()
```

```python
data={'first score':[100,90,np.nan,95],
      'second score':[30,45,56,np.nan],
      'third score':[np.nan,40,80,98]}
```

```python
data
```

```python
var=ps.DataFrame(data)
```

```python
var
```

```python
var.isnull()
var.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
imputer
```

```python
imputer.fit(var)
imputer.transform(var)
```
